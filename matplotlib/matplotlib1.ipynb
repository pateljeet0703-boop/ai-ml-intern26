```python
import matplotlib.pyplot as plt
from matplotlib import colors
```

```python



students = ["a","b","c","d","e"]
marks = [60,40,70,80,90]
plt.bar(students,marks)
plt.xlabel("students")
plt.ylabel("marrks")
plt.title("student marks")
for i in range (len(students)):
   plt.text(i,marks[i],str(marks[i]),ha="center",va="bottom")
plt.show()
```

```python
time=["11","12","1","2","3"]
orders =[100,120,40,50,60]
chinese =[12,18,3,6,9]
italian = [50,60,18,16,15]
plt.xlabel("time")
plt.ylabel("orders")
plt.title("zepto orderlist")
plt.bar(time,orders,label="orderlist")
plt.bar(time,chinese,bottom=orders,label="chinese")
plt.bar(time,italian,bottom=[orders[i]+chinese[i] for i in range(len(time))])
plt.show()
```
