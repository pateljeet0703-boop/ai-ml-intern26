```python
import cv2
import numpy as np

# Load image
img = cv2.imread("images.jpg")

# Resize for better collage fit
img = cv2.resize(img, (200, 200))

h, w = img.shape[:2]
center = (w // 2, h // 2)

images = []

# Generate 36 rotated images (0 to 350 step 10)
for angle in range(0, 360, 10):
    matrix = cv2.getRotationMatrix2D(center, angle, 1)
    rotated = cv2.warpAffine(img, matrix, (w, h))
    images.append(rotated)

# Convert list to 6x6 grid
rows = []
for i in range(0, 36, 6):  # 6 images per row
    row = np.hstack(images[i:i+6])
    rows.append(row)

final1_collage = np.vstack(rows)
final_collage = cv2.resize(final1_collage, (600, 600))  # Resize for better display


# Save and display
cv2.imwrite("36_angle_collage.jpg", final_collage)
cv2.imshow("36 Angle Collage", final_collage)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
