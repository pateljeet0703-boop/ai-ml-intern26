```python
import pandas as pd
import numpy as np
```

```python
# Create and Inspect: Create a DataFrame from the data dictionary. Print the specific shape of the DataFrame (rows, columns) and show the data types of all columns.
data = {
        "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace"],
    "Age": [25, 30, 35, 40, 22, 28, 45],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "San Antonio"],
    "Salary": [70000, 80000, 120000, 90000, 60000, 75000, 110000]
}
df = pd.DataFrame(data)
data
print(df.shape)
print(df.dtypes)
```

```python
# Drop Column: Remove the "City" column from the DataFrame.
data = {
        "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace"],
    "Age": [25, 30, 35, 40, 22, 28, 45],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "San Antonio"],
    "Salary": [70000, 80000, 120000, 90000, 60000, 75000, 110000]
}
df = pd.DataFrame(data)
df.drop(columns=["City"], inplace=True)
df
```

```python
# Drop Row: Remove the row at index 3.
df.drop(index=3, inplace=True)
print(df)
```

```python
# Add Constant Column: Add a new column "Country" and populate it with "USA" for all rows.
df["Country"] = "USA"
df
```

```python
# Add Calculated Column (Born Year): Assuming the current year is 2025, create a new column "Birth Year" by subtracting "Age" from 2025.
df["Birth Year"] = 2025 - df["Age"]
df
```

```python
# Add List Column: Add a new column "Department" using this list: ['HR', 'Eng', 'Eng', 'Sales', 'HR', 'Sales', 'Admin']
df["Department"] = ['HR', 'Eng', 'Eng', 'HR', 'Sales', 'Admin']
print(df)
```

```python
# Numpy conversion: Convert the underlying data of the DataFrame to a NumPy array and print it.
np_array = df.to_numpy()
print(np_array)
```

```python
# Empty DataFrame: Create an empty DataFrame. Then add two columns "A" and "B" with 3 values each.
df_empty = pd.DataFrame()
df_empty["A"] = [1, 2, 3]
df_empty["B"] = [4, 5, 6]
print(df_empty)
```

```python
# Reorder Columns: Rearrange the columns so that "Age" appears before "Name".
cols = list(df.columns)
cols.insert(0, cols.pop(cols.index("Age")))
df = df[cols]
print(df)
```

```python
# Conditional Update (Manual): Update the "Age" of "Alice" (index 0) to 26 using at or loc.
df.loc[0, "Age"] = 26
df
```
