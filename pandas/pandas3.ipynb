```python
import pandas as pd
```

```python
s = pd.Series([1,2,3,4,5])
print(s)
print(s.min)
print(s.max)
print(s.sum)
print(s.count)
print(s.var)
print(s.mean)
print(s.std)
```

```python
s = pd.Series([1,2,3,4,5,None,0,None,50])
print(s)

print(s.isnull().sum())
print(s.notnull().sum())

# print(s.fillna(0))
print(s.dropna())
```

```python
s = pd.Series([1,2,3,4,5,18,30,25,50])
print(s.sort_values())
print(s.sort_values(ascending=True))

print(s.sort_index())
print(s.sort_index(ascending=True))
```

```python
s = pd.Series([10,20,30,40])
print(s)

s1 = s.apply(lambda x:x*2)
s1
```

```python
s = pd.Series([10,20,30,40])
print(s)

s1 = s.map(lambda x:x*2)
s1
```

```python
s = pd.Series(['apple','banana','cherry',12])
print(s)

s = s.str.upper()
print(s)

s = s.str.lower()
print(s)

s = s.str.startswith("a")
print(s)
```
