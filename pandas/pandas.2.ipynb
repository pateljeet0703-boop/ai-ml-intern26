```python
import pandas as pd
```

```python
user = pd.Series([21,22,23,24,25],index=['xx','yy','zz','nn','ll'])
print(user.sample())
```

```python
print(user+1)
print(user-1)
print(user*1)
print(user/1)
```

```python
user1 = pd.Series([21,22,23,24,25],index=['xx','yy','zz','nn','ll'])
user2= pd.Series([21,22,23,24,25],index=['xx','yy','zz','nn','l'])
print(user1+user2)
```

```python
user11 = pd.Series([21,22,23,24,25],index=['xx','yy','zz','nn','ll'])
user22 = pd.Series([21,22,23,24,25],index=['xx','yy','zz','nn','l'])
result = (user11).add(user22,fill_value=0)
print(result)
result1 = (user11).sub(user22,fill_value=0)
print(result1)
result2 = (user11).mul(user22,fill_value=0)
print(result2)
result3 = (user11).div(user22,fill_value=0)
print(result3)
```
