```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np  
# import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('battery_drain_clean_data (1).csv')
```

```python
data.isnull().sum()
```

```python
data.sample(5)
```

```python
x = data[["Social_Media_Hours"]]
y = data[["Battery_Drain_Percentage"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python

model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
compare = pd.DataFrame({"Actual":y_test.values.flatten(),"Predicted":y_pred.flatten()}) 
print(compare)
```

```python
plt.scatter(x_train,y_train,color="blue")
plt.plot(x_train,model.predict(x_train),color="red")
plt.title("Battery Drain vs Social Media Hours (Training Set)")
plt.xlabel("Social Media Hours")
plt.ylabel("Battery Drain Percentage")
plt.show()
```
